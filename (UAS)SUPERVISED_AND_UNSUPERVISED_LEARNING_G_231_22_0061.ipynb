```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Generate data secara acak
np.random.seed(0)
X = 2.5 * np.random.randn(100) + 1.5   # features
res = 0.5 * np.random.randn(100)       # error
y = 2 + 0.3 * X + res                  # target

# Membagi data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Membentuk model regresi linear
model = LinearRegression()
model.fit(X_train.reshape(-1, 1), y_train)

# Memprediksi nilai dengan model yang sudah dilatih
y_pred = model.predict(X_test.reshape(-1, 1))

# Plot hasil prediksi
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.title('Supervised Learning: Regresi Linear')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate data secara acak
np.random.seed(0)
X = np.random.rand(100, 2)

# Membentuk model K-Means dengan 3 kelompok (clusters)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Mendapatkan label dari kluster yang terbentuk
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_

# Plot hasil klustering
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red', label='Centroids')
plt.title('Unsupervised Learning: K-Means Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
